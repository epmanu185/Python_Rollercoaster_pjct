## 4. Ask questions about the data
         - what are the locations with fastest roller coasters (minium 10)?

```python
df.head()
```

```python
df['Location'].value_counts()
```

```python
df.query('Location != "others"').groupby('Location')['Speed_mph'].agg(['mean','count']).query('count >= 10').sort_values('mean')
```

```python
ax = df.query('Location != "others"').groupby('Location')['Speed_mph'].agg(['mean','count']).query('count >= 10').sort_values('mean')['mean']\
.plot(kind='barh', figsize=(12,5), title='Avvg coaster speed by location')
```
